```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/kaggle/input/sample-sales-data/sales_data_sample.csv' ,encoding='latin1')


plt.figure(figsize=(10, 5))
sns.lineplot(data=df, x='ORDERDATE', y='SALES') 
plt.title('Sales Over Time')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(data=df, x='PRODUCTLINE', y='SALES') 
plt.title('Sales by Product')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(data=df, x='PRICEEACH', y='SALES') # There is no profit so it has been replaced with PRICE EACH column
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x='COUNTRY', y='SALES', estimator=sum, ci=None)  
plt.title('Total Sales by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)  
plt.show()
```

```python
sales_by_region = df.groupby('COUNTRY')['SALES'].sum() 
plt.figure(figsize=(8, 8))
sales_by_region.plot.pie(autopct='%1.1f%%')  
plt.title('Percentage of Sales by Region')
plt.ylabel('')  
plt.show()
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt

world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

world = world.merge(df, how="left", left_on="name", right_on="COUNTRY")

fig, ax = plt.subplots(1, 1, figsize=(15, 10))
world.boundary.plot(ax=ax) 
world.plot(column='SALES', ax=ax, legend=True,
           legend_kwds={'label': "Sales by Country",
                        'orientation': "horizontal"},
           cmap='viridis')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/kaggle/input/sample-sales-data/sales_data_sample.csv',encoding='latin1')

sales_by_product = df.groupby('PRODUCTLINE')['SALES'].sum()

sorted_sales = sales_by_product.sort_values(ascending=False)

top_products = sorted_sales.head(5)

total_sales = df['SALES'].sum()

percentage_top_products = (top_products / total_sales) * 100

print("Percentage of Sales by Top-Performing Products:")
print(percentage_top_products)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/kaggle/input/sample-sales-data/sales_data_sample.csv',encoding='latin1')

grouped = df.groupby('COUNTRY')
regional_sales = grouped['SALES'].sum()

regional_performance = pd.DataFrame({
    'Total Sales': regional_sales
})
regional_performance = regional_performance.sort_values(by='Total Sales', ascending=False)

print("Regional Performance Analysis:")
print(regional_performance)
```
